## rollback
이전 Production 버전을 다시 활성화(메타데이터 전환)

```python
from mlflow.tracking import MlflowClient

MODEL_NAME = "team4-pred-models"
client = MlflowClient()

# 문제가 생기면 Rollback (예: v1 다시 Production)
client.transition_model_version_stage(
    name=MODEL_NAME,
    version=1,
    stage="Production",
    archive_existing_versions=True
)
```
