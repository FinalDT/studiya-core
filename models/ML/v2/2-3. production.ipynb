## production
운영용(서비스/엔드포인트가 Production 버전을 가리키면 실제 트래픽에 사용됨)

```python
from mlflow.tracking import MlflowClient

MODEL_NAME = "team4-pred-models"
client = MlflowClient()

# 테스트 완료 후 Production으로 전환
client.transition_model_version_stage(
    name=MODEL_NAME,
    version=2,
    stage="Production",
    archive_existing_versions=True   # 이전 Production 자동 archive
)
```
